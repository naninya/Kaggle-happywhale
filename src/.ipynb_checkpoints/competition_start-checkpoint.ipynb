# Unzip data

```python
data_path = "./data"
img_path = os.path.join(data_path, "img")
submission_file_path = os.path.join(data_path, "sample_submission.csv.zip")
train_csv_file_path = os.path.join(data_path, "train.csv.zip")
!unzip -uq $submission_file_path -d ./data
!unzip -uq $train_csv_file_path -d ./data
submission_file_path = ".".join(submission_file_path.split(".")[:-1])
train_csv_file_path = ".".join(train_csv_file_path.split(".")[:-1])

img_list = os.listdir(img_path)
zip_img_path_list = []
for img_file_name in img_list:
    file_path = os.path.join(img_path, img_file_name)
    if ".zip" in file_path:
        zip_img_path_list.append(file_path)

for zip_file_path in zip_img_path_list:
    !unzip -uq $zip_file_path -d ./data/img
img_path_dict = {}
for img_file_name in os.listdir(img_path):
    file_path = os.path.join(img_path, img_file_name)
    if ".zip" not in file_path:
        img_path_dict[img_file_name] = file_path
```

```python
train = pd.read_csv(train_csv_file_path)
train["species"] = train["species"].replace(["bottlenose_dolpin", "kiler_whale",
                                             "beluga", 
                                             "globis", "pilot_whale"],
                                            ["bottlenose_dolphin", "killer_whale",
                                             "beluga_whale", 
                                             "short_finned_pilot_whale", "short_finned_pilot_whale"])
train
```

```python
train["is_img_file"] = False
train["img_file_path"] = None
train.loc[train["image"].isin(img_path_dict), "is_img_file"] = True
train["img_file_path"] = train["image"].map(img_path_dict)
train
```

```python
import cv2
img_show_sample = train[train["is_img_file"]].copy()
img_show_sample = img_show_sample.sample(n=min(5, img_show_sample.shape[0]), random_state=0)
fig = plt.figure(figsize=(18, 5 * 5))
for index, img_path in enumerate(img_show_sample["img_file_path"]):
    data = cv2.imread(img_path)
    ax = fig.add_subplot(5, 1, index + 1)
    ax.imshow(data)
```
