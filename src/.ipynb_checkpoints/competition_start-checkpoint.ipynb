# Import

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
WORK_DIR = r"C:\Users\dnsgu\Desktop\workspace\git_project\Happywhale"
os.chdir(WORK_DIR)
```

# Unzip data

```python
data_path = "./data"
img_path = os.path.join(data_path, "img")
submission_file_path = os.path.join(data_path, "sample_submission.csv.zip")
train_csv_file_path = os.path.join(data_path, "train.csv.zip")
!unzip -uq $submission_file_path -d ./data
!unzip -uq $train_csv_file_path -d ./data
submission_file_path = ".".join(submission_file_path.split(".")[:-1])
train_csv_file_path = ".".join(train_csv_file_path.split(".")[:-1])

img_list = os.listdir(img_path)
zip_img_path_list = []
for img_file_name in img_list:
    file_path = os.path.join(img_path, img_file_name)
    if ".zip" in file_path:
        zip_img_path_list.append(file_path)

for zip_file_path in zip_img_path_list:
    !unzip -uq $zip_file_path -d ./data/img
img_path_dict = {}
for img_file_name in os.listdir(img_path):
    file_path = os.path.join(img_path, img_file_name)
    if ".zip" not in file_path:
        img_path_dict[img_file_name] = file_path
```

# Correct name missmatching

```python
train = pd.read_csv(train_csv_file_path)
train["species"] = train["species"].replace(["bottlenose_dolpin", "kiler_whale",
                                             "beluga", 
                                             "globis", "pilot_whale"],
                                            ["bottlenose_dolphin", "killer_whale",
                                             "beluga_whale", 
                                             "short_finned_pilot_whale", "short_finned_pilot_whale"])
train
```

# Add cols : 
* is_img_file
* img_file_path

```python
train["is_img_file"] = False
train["img_file_path"] = None
train.loc[train["image"].isin(img_path_dict), "is_img_file"] = True
train["img_file_path"] = train["image"].map(img_path_dict)
train
```

# Add cols:
*category

```python
train["category"] = train["species"].map(lambda x:x.split("_")[-1])
train
```

# Check image sample

```python
import cv2
img_show_sample = train[train["is_img_file"]].copy()
img_show_sample = img_show_sample.sample(n=min(5, img_show_sample.shape[0]), random_state=0)
fig = plt.figure(figsize=(18, 5 * 5))
for index, img_path in enumerate(img_show_sample["img_file_path"]):
    data = cv2.imread(img_path)
    ax = fig.add_subplot(5, 1, index + 1)
    ax.imshow(data)
```

# Data details

```python
# the number of whales
train_img_num = train.shape[0]
train_whale_num = train["individual_id"].unique().shape[0]
train_species_num = train["species"].unique().shape[0]

print(f"train_img_num: {train_img_num}")
print(f"train_whale_num: {train_whale_num}")
print(f"train_species_num: {train_species_num}")
```

# Plotting species

```python
fig = plt.figure(figsize=(18,5))
ax = fig.add_subplot(1,2,1)
ax.hist(train["species"], color="green")
ax.set_title("species_cnt")
ax = fig.add_subplot(1,2,2)
cnt_species = train["species"].value_counts()
ax.pie(
    cnt_species,
    labels = cnt_species.index,
    autopct='%1.0f%%',
    counterclock=False,
    startangle=90
)
plt.show()
```

# Plotting categories

```python
fig = plt.figure(figsize=(18,5))
ax = fig.add_subplot(1,2,1)
ax.hist(train["category"], color="green")
ax.set_title("category")
ax = fig.add_subplot(1,2,2)
cnt_species = train["category"].value_counts()
ax.pie(
    cnt_species,
    labels = cnt_species.index,
    autopct='%1.0f%%',
    counterclock=False,
    startangle=90
)
plt.show()
```

# plotting number of fig on unique id

```python
import numpy as np
fig = plt.figure(figsize=(18,5))
ax = fig.add_subplot(1,1,1)
cnt_id = train["individual_id"].value_counts()
print(train["individual_id"].value_counts())
bins_num = int(1 + np.ceil(np.log2(train["individual_id"].shape[0])))
ax.hist(
    cnt_id,
    bins=bins_num
)
plt.show()
```
